# Métodos Runge–Kutta **implícitos**

<span style="font-family:Georgia;">

Dado el problema de valor inicial
$$
y'(t) = f(t,y(t)),\qquad y(t_0)=y_0,\qquad y:\mathbb{R}\to\mathbb{R}^m,
$$
un método Runge–Kutta de $s$ etapas queda determinado por su **tabla de Butcher**
$$
\begin{array}{c|ccc}
c_1 & a_{11} & \cdots & a_{1s} \\
\vdots & \vdots & \ddots & \vdots \\
c_s & a_{s1} & \cdots & a_{ss} \\
\hline
 & b_1 & \cdots & b_s
\end{array}
\quad\text{con}\quad c = A\mathbf{1}\;(\text{es decir }c_i=\sum_j a_{ij}),\; A=(a_{ij}).
$$

Las **etapas** $K_i$ se definen por
$$
K_i \;=\; f\!\Big(t_n + c_i h,\; y_n + h\sum_{j=1}^{s} a_{ij} K_j\Big),\qquad i=1,\dots,s,
$$
y el **paso** es
$$
y_{n+1} \;=\; y_n + h\sum_{i=1}^s b_i K_i.
$$

Decimos que el método es **implícito** si $A$ no es estrictamente triangular superior (las $K_i$ aparecen en ambos lados). En cada paso hay que resolver el sistema no lineal en las etapas $K\in\mathbb{R}^{sm}$:
$$
G(K) = 0,\qquad 
G_i(K) := K_i - f\!\Big(t_n + c_i h,\; y_n + h\sum_{j=1}^{s} a_{ij} K_j\Big).
$$

## Métodos RK **implícitos** de 2 etapas
Consideremos $s=2$ con parámetros
$$
A=\begin{bmatrix}a_{11}&a_{12}\\ a_{21}&a_{22}\end{bmatrix},\quad
b=\begin{bmatrix}b_1\\ b_2\end{bmatrix},\quad
c=\begin{bmatrix}c_1\\ c_2\end{bmatrix}.
$$

Denotemos
$$
F := f(t_n,y_n)
$$
La **derivada total** a lo largo de la solución exacta $y(t)$ es
$$
D := \frac{d}{dt} = \partial_t + f_y(t,y)\,[\,\cdot\,].
$$

Al hacer la expansión de Taylor de la solución obtenemos
$$
y(t_n+h) = y_n + h F + \tfrac{h^2}{2} D F + \tfrac{h^3}{6} D^2 F + \tfrac{h^4}{24} D^3 F + O(h^5),
$$
emparejamos coeficientes de $A,b$ y $c$ en las expansiones de Taylor de las etapas $K_i$ del método RK y deducimos las condiciones de orden.

###  **Condiciones de orden** 

- **Orden 1**
$$
\boxed{\;\sum_{i=1}^2 b_i = 1\;}
$$

- **Orden 2**
$$
\boxed{\;\sum_{i=1}^2 b_i c_i = \frac{1}{2}\;}
$$

- **Orden 3** 
$$
\boxed{\;\sum_{i=1}^2\sum_{j=1}^2 b_i a_{ij} c_j = \frac{1}{6}\;}
$$

$$
\boxed{\;\sum_{i=1}^2 b_i c_i^2 = \frac{1}{3}\;}
$$

- **Orden 4**
$$
\boxed{\;\sum_i b_i c_i^3 = \frac{1}{4}\;},\qquad
\boxed{\;\sum_{i,j} b_i\,c_i\,a_{ij} c_j = \frac{1}{8}\;},\qquad
\boxed{\;\sum_{i,j} b_i\,a_{ij} c_j^2 = \frac{1}{12}\;},\qquad
\boxed{\;\sum_{i,j,k} b_i\,a_{ij} a_{jk} c_k = \frac{1}{24}\;}
$$

###  Dos métodos RK2 implícitos canónicos

- **Radau IIA (2 etapas)** — orden $p=3$:
$$
\begin{array}{c|cc}
\frac{1}{3} & \frac{5}{12} & -\frac{1}{12}\\[2pt]
1 & \frac{3}{4} & \frac{1}{4}\\[3pt]
\hline
 & \frac{3}{4} & \frac{1}{4}
\end{array}
$$

**Etapas:**

$$
\begin{aligned}
K_1 &= f\!\Big(t_n+\tfrac{h}{3},\; y_n + h\big(\tfrac{5}{12}K_1 - \tfrac{1}{12}K_2\big)\Big),\\[4pt]
K_2 &= f\!\Big(t_n+h,\; y_n + h\big(\tfrac{3}{4}K_1 + \tfrac{1}{4}K_2\big)\Big).
\end{aligned}
$$

**Paso:**

$$
y_{n+1} = y_n + h\Big(\tfrac{3}{4}K_1 + \tfrac{1}{4}K_2\Big).
$$

- **Gauss–Legendre (2 etapas)** — orden $p=4$:
$$
\begin{array}{c|cc}
\frac{1}{2}-\frac{\sqrt{3}}{6} & \frac{1}{4} & \frac{1}{4}-\frac{\sqrt{3}}{6}\\[2pt]
\frac{1}{2}+\frac{\sqrt{3}}{6} & \frac{1}{4}+\frac{\sqrt{3}}{6} & \frac{1}{4}\\[3pt]
\hline
 & \frac{1}{2} & \frac{1}{2}
\end{array}
$$

**Etapas:**

$$
\begin{aligned}
K_1 &= f\!\Big(t_n+\big(\tfrac{1}{2}-\tfrac{\sqrt{3}}{6}\big)h,\;
y_n + h\big(\tfrac{1}{4}K_1 + \big(\tfrac{1}{4}-\tfrac{\sqrt{3}}{6}\big)K_2\big)\Big),\\[4pt]
K_2 &= f\!\Big(t_n+\big(\tfrac{1}{2}+\tfrac{\sqrt{3}}{6}\big)h,\;
y_n + h\big(\big(\tfrac{1}{4}+\tfrac{\sqrt{3}}{6}\big)K_1 + \tfrac{1}{4}K_2\big)\Big).
\end{aligned}
$$

**Paso:**

$$
y_{n+1} = y_n + \tfrac{h}{2}\,(K_1 + K_2).
$$

```python
import sympy as sp

# Radau IIA(2)
A_r = sp.Matrix([[sp.Rational(5,12), -sp.Rational(1,12)],
                 [sp.Rational(3,4), sp.Rational(1,4)]])
b_r = sp.Matrix([sp.Rational(3,4), sp.Rational(1,4)])
c_r = sp.Matrix([sp.Rational(1,3), sp.Integer(1)])

# Gauss-Legendre(2)
sqrt3 = sp.sqrt(3)
A_g = sp.Matrix([[sp.Rational(1,4), sp.Rational(1,4)-sqrt3/6],
                  [sp.Rational(1,4)+sqrt3/6, sp.Rational(1,4)]])
b_g = sp.Matrix([sp.Rational(1,2), sp.Rational(1,2)])
c_g = sp.Matrix([sp.Rational(1,2)-sqrt3/6, sp.Rational(1,2)+sqrt3/6])

def verificar_condiciones(A,b,c, name):
    oc1 = sp.simplify(sum(b))                                   # sum b_i
    oc2 = sp.simplify(sum(b[i]*c[i] for i in range(2)))         # sum b_i c_i
    oc3a = sp.simplify(sum(b[i]*c[i]**2 for i in range(2)))     # sum b_i c_i^2
    oc3b = sp.nsimplify(sum(b[i]*A[i,j]*c[j] for i in range(2) for j in range(2)))  # sum b_i a_ij c_j
    oc4a = sp.simplify(sum(b[i]*c[i]**3 for i in range(2)))
    oc4b = sp.nsimplify(sum(b[i]*c[i]*A[i,j]*c[j] for i in range(2) for j in range(2)))
    oc4c = sp.nsimplify(sum(b[i]*A[i,j]*c[j]**2 for i in range(2) for j in range(2)))
    oc4d = sp.nsimplify(sum(b[i]*A[i,j]*A[j,k]*c[k] for i in range(2) for j in range(2) for k in range(2)))
    print(f"== {name} ==")
    print("sum b_i = ", oc1, " (valor teórico 1)")
    print("sum b_i c_i = ", oc2, " (valor teórico 1/2)")
    print("sum b_i c_i^2 = ", oc3a, " (valor teórico 1/3)")
    print("sum_{i,j} b_i a_ij c_j = ", oc3b, " (valor teórico 1/6)")
    print("sum b_i c_i^3 = ", oc4a, " (valor teórico 1/4)")
    print("sum b_i c_i a_ij c_j = ", oc4b, " (valor teórico 1/8)")
    print("sum b_i a_ij c_j^2 = ", oc4c, " (valor teórico 1/12)")
    print("sum b_i a_ij a_jk c_k = ", oc4d, " (valor teórico 1/24)")
    print()

verificar_condiciones(A_r,b_r,c_r, "Radau IIA(2)  (orden 3)")
verificar_condiciones(A_g,b_g,c_g, "Gauss-Legendre(2)  (orden 4)")
```

# Modelo de **Michaelis–Menten**

A comienzos del siglo XX, los bioquímicos **Leonor Michaelis** y **Maud Menten** desarrollaron un modelo matemático para describir cómo las **enzimas** catalizan reacciones químicas.  
Una enzima ($E$) es una proteína que acelera la conversión de un **sustrato** ($S$) en un **producto** ($P$). En el proceso, la enzima se combina temporalmente con el sustrato para formar un **complejo enzima–sustrato** ($ES$).  
El modelo más simple que describe esta reacción es:

$$
E + S \;\xrightleftharpoons[k_{-1}]{k_1}\; ES \;\xrightarrow{k_2}\; E + P.
$$

donde:
- $k_1$ es la constante de velocidad de **asociación** (formación de $ES$),
- $k_{-1}$ es la constante de **disociación** de $ES$ hacia $E+S$,
- $k_2$ es la constante de **catálisis** que transforma $ES$ en producto $P$ liberando la enzima.



### Sistema de ecuaciones de la reacción

Denotemos las concentraciones:
$$
e = [E],\qquad s = [S],\qquad c = [ES],\qquad p = [P].
$$
Por la **ley de acción de masas**, el sistema dinámico es:
$$
\begin{aligned}
\dot{e} &= -k_1 e s + (k_{-1} + k_2)c,\\
\dot{s} &= -k_1 e s + k_{-1} c,\\
\dot{c} &= k_1 e s - (k_{-1} + k_2)c,\\
\dot{p} &= k_2 c.
\end{aligned}
$$

Si sumamos las ecuaciones de $\dot{e}$ y $\dot{c}$ obtenemos:
$$
\dot{e} + \dot{c} = \big(-k_1 e s + (k_{-1}+k_2)c\big) + \big(k_1 e s - (k_{-1}+k_2)c\big) = 0,
$$
por lo que $e + c = E_0$ es **constante**.  
De aquí $e = E_0 - c$, lo que permite reducir el sistema a tres variables $(s, c, p)$:
$$
\begin{aligned}
\dot{s} &= -k_1 (E_0 - c) s + k_{-1} c,\\
\dot{c} &= k_1 (E_0 - c)s - (k_{-1}+k_2)c,\\
\dot{p} &= k_2 c.
\end{aligned}
$$



### Aproximación de Michaelis–Menten (Estado estacionario)

Si la concentración total de enzima es pequeña ($E_0 \ll s_0$), el complejo $ES$ alcanza rápidamente un equilibrio cuasi–estacionario.  
Suponemos entonces que $\dot{c}\approx 0$, es decir, que la **concentración de $ES$ cambia muy lentamente** comparada con la del sustrato.  
De $\dot{c}\approx0$ se deduce:
$$
0 \approx k_1 (E_0 - c)s - (k_{-1}+k_2)c,
$$
de donde
$$
c \approx \frac{E_0 s}{K_M + s}, \qquad K_M := \frac{k_{-1}+k_2}{k_1}.
$$
Este supuesto significa que la formación y descomposición del complejo $ES$ están **en equilibrio dinámico**, y la reacción global está limitada por la etapa catalítica controlada por $k_2$.

Definiendo $V_{\max} = k_2 E_0$, las ecuaciones reducidas son:
$$
\dot{p} = \frac{V_{\max} s}{K_M + s}, \qquad \dot{s} = -\frac{V_{\max} s}{K_M + s}.
$$


### Ley **saturante** y su interpretación

La velocidad de reacción del producto $p$ es
$$
v(s) = \dot{p} = \frac{V_{\max} s}{K_M + s}.
$$

- Para **baja concentración** de sustrato ($s \ll K_M$):
  $$v(s) \approx \frac{V_{\max}}{K_M} s,$$
  es decir, la velocidad crece **linealmente** con $s$ (régimen de primer orden).

- Para **alta concentración** de sustrato ($s \gg K_M$):
  $$v(s) \approx V_{\max},$$
  alcanzando un **valor máximo** independiente de $s$ (régimen de saturación).

- En $s = K_M$:
  $$v(K_M) = \frac{V_{\max}}{2},$$
  por lo que $K_M$ se interpreta como la **concentración a la cual la enzima trabaja a la mitad de su velocidad máxima**.
